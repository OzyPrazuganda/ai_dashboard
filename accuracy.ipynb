```python
def calculate_accuracy(df):
    date = df['day'].iloc[0]
    df_filtered = df[df.iloc[:, 21] == "是"].iloc[:, [21, 34]]
    denom = (df_filtered.iloc[:,0] == "是").sum()
    if denom == 0:
        return np.nan, date

    accuracy = (df_filtered.iloc[:, 1] == "是").sum() / denom * 100
    return accuracy, date

dataframes = [df, df_22, df_28, df_8_5, df_8_11, df_8_18, df_8_27, df_9_2, df_9_15, df_9_29, df_10_6, df_10_13, df_10_20, df_10_27, df_11_3, df_11_10, df_11_16,
 df_11_17, df_11_18, df_11_24]
results = []

for  data in dataframes:
    ratio, date = calculate_accuracy(data)
    results.append({
        'date': date,
        'ratio': ratio
    })

accuracy_history = pd.DataFrame(results)
accuracy_history
```

```python
accuracy_history['date'] = pd.to_datetime(accuracy_history['date'])

#visualize using matplotlib
plt.figure(figsize=(12, 6))
plt.plot(accuracy_history['date'], accuracy_history['ratio'], marker='o')
plt.title('Accuracy Over Time')
plt.xlabel('Date')
plt.ylabel('Accuracy (%)')
plt.grid(True)
plt.tight_layout()

plt.show()
```
