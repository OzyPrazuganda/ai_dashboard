```python
import pandas as pd
import pygwalker as pyg
```

```python
df = pd.read_csv('dataset_accuracy/akulaku_superset_presto_hanmm_ozy.prazuganda_1753329431_21.csv')
df_22 = pd.read_csv('dataset_accuracy/akulaku_superset_presto_hanmm_ozy.prazuganda_1753341756_22.csv')
df_28 = pd.read_csv('dataset_accuracy/akulaku_superset_presto_hanmm_ozy.prazuganda_1753775670_28.csv')
df_8_5 = pd.read_csv('dataset_accuracy/akulaku_superset_presto_hanmm_ozy.prazuganda_1754385922_8_5.csv')
df_8_11 = pd.read_csv('dataset_accuracy/akulaku_superset_presto_hanmm_ozy.prazuganda_1755661415_8_11.csv')
df_8_18 = pd.read_csv('dataset_accuracy/akulaku_superset_presto_hanmm_ozy.prazuganda_1755661737_8_18.csv')
```

```python
pyg.walk(df)
```

```python
def calculate_accuracy(df):
    date = df['day'].unique()
    df_filtered = df[df.iloc[:, 21] == "是"].iloc[:, [21, 34]]
    accuracy = (df_filtered.iloc[:, 1] == "是").sum() / (df_filtered.iloc[:, 0] == "是").sum() * 100
    return accuracy, date

dataframes = [df, df_22, df_28, df_8_5, df_8_11, df_8_18]

for i, data in enumerate(dataframes, start=1):
    ratio, date = calculate_accuracy(data)
    print(f"{ratio:.2f} % Accuracy per {date}")
```
