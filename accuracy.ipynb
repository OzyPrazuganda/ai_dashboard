```python
def calculate_accuracy(df):
    date = df['day'].unique()
    df_filtered = df[df.iloc[:, 21] == "是"].iloc[:, [21, 34]]
    accuracy = (df_filtered.iloc[:, 1] == "是").sum() / (df_filtered.iloc[:, 0] == "是").sum() * 100
    return accuracy, date

dataframes = [df, df_22, df_28, df_8_5, df_8_11, df_8_18, df_8_27, df_9_2, df_9_15, df_9_29, df_10_6, df_10_13]

for i, data in enumerate(dataframes, start=1):
    ratio, date = calculate_accuracy(data)
    print(f"{ratio:.2f} % Accuracy per {date}")
```
