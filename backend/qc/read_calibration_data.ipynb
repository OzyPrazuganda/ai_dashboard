```python
import os
import gspread
import pandas as pd
from IPython.display import Audio
from google.oauth2.service_account import Credentials
```

```python
# Read data from gsheet

SERVICE_ACCOUNT_FILE = '../../key/credentials.json'
SCOPES = ['https://www.googleapis.com/auth/spreadsheets',
          'https://www.googleapis.com/auth/drive']
creds = Credentials.from_service_account_file(
    SERVICE_ACCOUNT_FILE, scopes=SCOPES)
client = gspread.authorize(creds)

sheet = client.open("Kalibrasi & Standarisasi AI QC 统一规范&异常讨论").worksheet("Kalib Sampling")
data = sheet.get_all_values()

df = pd.DataFrame(data[192:], columns=data[0])
```

```python
# clean data
df_clean = df.fillna("")
df_clean = df_clean.drop(df_clean.columns[[0,6,7,10,12,13,14,15,16,17,18,19]], axis=1)
df_clean["Tanggal Meeting"] = pd.to_datetime(df_clean["Tanggal Meeting"], errors="coerce").dt.date

df_clean.to_csv("../../dataset_qc/weekly_calibration_data.csv")
```
